# Accumulate and visualize the hyperparameter tunning 

```python
import json 
import pandas as pd 
import os 
import numpy as np 
import pickle
```

```python

# accumulate and vizualize the results of the experiment.
root = '/Volumes/cardio-project/cardio/SPUM/CVD_detection_code/cnn/results/tuning_augment'
config_dict = json.load( open( os.path.join(root,'config_prameter_tuning.json') ))

line = [] # exp, lr, wd, drp, best F1 mean,best F1 std, epochs 
for key , dict in config_dict.items():
    out = []
    print(key)
    accumulate_f1 = []
    for file_name in dict['paths']:
        with open(os.path.join(root, file_name ), "rb") as fp:
            perf_tmp = pickle.load(fp)
        accumulate_f1.append(perf_tmp['f1_score'])

    f1_mean = np.mean(accumulate_f1,axis = 0)
    f1_std = np.std(accumulate_f1,axis = 0)
    best_epochs = np.argmax(f1_mean) + 1
    best_f1_mean = f1_mean[np.argmax(f1_mean)]
    best_f1_std = f1_std[np.argmax(f1_mean)]
    out.append(key)
    out.append(dict['lr'])
    out.append(dict['wd'])
    out.append(dict['drp'])
    out.append(best_f1_mean)
    out.append(best_f1_std)
    out.append(best_epochs)
    line.append(out)

results = pd.DataFrame(line, columns=['exp', 'lr', 'wd', 'drp', 'best F1 mean','best F1 std', 'epoch'])
results = results.sort_values(by=['best F1 mean'])
```

```python
from sklearn.metrics import confusion_matrix
# accumulate and vizualize the results of the test
root = '/Volumes/cardio-project/cardio/SPUM/CVD_detection_code/cnn/results/testing'


line_scratch = [] #  best F1 mean,best F1 std, epochs 
line_img = []
line_best = []
names = ['augmented_culprit']
for name in names : 
    name_files = { 'Scratch': [ name+'_perf_scratch'+'_run_'+str(k+1)+'.pkl' for k in range(5)] ,
                    'Fine-tune imgNet':[ name+'_perf_ft_imgnet'+'_run_'+str(k+1)+'.pkl' for k in range(5)] , 
                    'Fine-tune best': [ name+'_perf_ft_best'+'_run_'+str(k+1)+'.pkl' for k in range(5)]}
    
    

    for k, namefiles in name_files.items():

        print('---------',k)
        accumulate = {
        'f1_score': [],
        'recall': [],
        'specificity': [],
        'precision': [],
        'auc': []
        }
        accumulate_std = {
        'f1_score': [],
        'recall': [],
        'specificity': [],
        'precision': [],
        'auc': []
        }
        for name_file in namefiles:
            with open(os.path.join(root, name_file ), "rb") as fp:
                perf_tmp = pickle.load(fp)
            accumulate['f1_score'].append(perf_tmp['f1_score'])
            accumulate['recall'].append(perf_tmp['recall'])
            accumulate['specificity'].append(perf_tmp['specificity'])
            accumulate['precision'].append(perf_tmp['precision'])
            accumulate['auc'].append(perf_tmp['auc'])

        for k,v in accumulate.items():
            accumulate[k] = np.mean(v ,axis=0)
            accumulate_std[k] = 2*np.std(v ,axis=0)

        best_epochs = np.argmax(accumulate['f1_score'])
        print('best epochs',best_epochs )
        for k,v in accumulate.items():
            print(k,v[best_epochs])
            print(k+' std',accumulate_std[k][best_epochs])
        
        print('confusion')
        print(confusion_matrix(perf_tmp['y_true'],perf_tmp['y_pred']))

        
        
        
```
